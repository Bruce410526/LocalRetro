```python
from rdchiral.template_extractor import extract_from_reaction
from SimpleChiral import reduce_template

def extract_templates(rxns):
    simple_templates = defaultdict(int)
    train_templates = []
    for i, rxn in enumerate(rxns):
        if True:
            rxn = {'reactants': rxn.split('>>')[0], 'products': rxn.split('>>')[1], '_id': i}
            result = extract_from_reaction(rxn, local = True)
            if 'reaction_smarts' not in result.keys():
                continue
            local_template = result['reaction_smarts']
            simple_template = reduce_template(local_template)
            simple_templates[simple_template] += 1
    return simple_templates
```

```python
from Run_preprocessing import get_label
import pandas as pd

def get_nums(a):
    a = str(a)
    nums = []
    c = ''
    for s in a:
        if s.isdigit():
            c += s
        else:
            if c != '':
                nums.append(int(c))
                c = ''
    if c != '':
        nums.append(int(c))
    return sorted(nums)

data = pd.read_csv('../data/USPTO_50K/raw_train.csv')['reactants>reagents>production']
preprocessed = pd.read_csv('../data/USPTO_50K/preprocessed_train.csv')['edit_idx']
different = []
for n, rxn in enumerate(data):
    rxn, template = get_label(rxn)
    if 'edit_sites' not in template.keys():
        continue
    inum = template['edit_sites'][0]
    if len(inum) == 1:
        if inum[0] != eval(preprocessed[n]):
            if type(inum[0]) == type(0):
                if inum[0] != eval(preprocessed[n]):
                    different.append(n)
                    print (n, inum, eval(preprocessed[n]))
            elif get_nums(inum) != get_nums(preprocessed[n]):  
                different.append(n)
                print (n, inum, eval(preprocessed[n]))
    else:
        a = get_nums(inum)
        b = get_nums(preprocessed[n])
        if  a != b:
            if a == []:
                print (n ,rxn, inum)
            different.append(n)
            print (n, inum, preprocessed[n])
        
    if n % 500 == 0:
        print (n, len(different))
#     if n == 134:
#         break
```

```python
def check_break_bond(rxn):
    product = rxn.split('>>')[1]
    reactant = rxn.split('>>')[0]
    product = Chem.MolFromSmiles(product)
    reactant = Chem.MolFromSmiles(reactant)
    product_bonds = []
    atom_map = {atom.GetAtomMapNum(): atom.GetIdx() for atom in product.GetAtoms()}
    for bond in product.GetBonds():
        atoma = bond.GetBeginAtom().GetAtomMapNum()
        atomb = bond.GetEndAtom().GetAtomMapNum()
        bond_map = (min([atoma, atomb]), max([atoma, atomb]))
        product_bonds.append(bond_map)
    
    break_bonds_map = None
    break_bonds_idx = None
    rxn_center = None
    for bond in reactant.GetBonds():
        atoma = bond.GetBeginAtom().GetAtomMapNum()
        atomb = bond.GetEndAtom().GetAtomMapNum()
        bond_map = (min([atoma, atomb]), max([atoma, atomb]))
        if bond_map[0] != 0 and bond_map[1] != 0 and bond_map not in product_bonds:
            break_bonds_map = bond_map
            break_bonds_idx = [atom_map[num] for num in bond_map]
            rxn_center = '*'.join(sorted([bond.GetBeginAtom().GetSymbol(), bond.GetEndAtom().GetSymbol()]))
    return rxn_center, break_bonds_map, break_bonds_idx

check_break_bond(rxn)
```

```python
bond_type_map = {'SINGLE': '-', 'DOUBLE': '=', 'TRIPLE': '#', 'AROMATIC': '@'}

def check_change_aromatic(rxn):
    product = rxn.split('>>')[1]
    reactant = rxn.split('>>')[0]
    product = Chem.MolFromSmiles(product)
    reactant = Chem.MolFromSmiles(reactant)
    bond_map_bond = {}
    atom_map = {atom.GetAtomMapNum(): atom.GetIdx() for atom in product.GetAtoms()}
    for bond in product.GetBonds():
        atoma = bond.GetBeginAtom().GetAtomMapNum()
        atomb = bond.GetEndAtom().GetAtomMapNum()
        bond_map = (min([atoma, atomb]), max([atoma, atomb]))
        bond_map_bond[bond_map] = bond
    
    change_idx = 1
    change_bonds_map = {}
    change_bonds_idx = {}
    rxn_center = {}
    for bond in reactant.GetBonds():
        atoma = bond.GetBeginAtom().GetAtomMapNum()
        atomb = bond.GetEndAtom().GetAtomMapNum()
        bond_map = (min([atoma, atomb]), max([atoma, atomb]))
        if bond_map in bond_map_bond.keys():
            if bond_map_bond[bond_map].GetIsAromatic() != bond.GetIsAromatic():
                change_bonds_map[change_idx] = bond_map
                change_bonds_idx[change_idx] = [atom_map[num] for num in bond_map]
                bond_type = bond_type_map[str(bond.GetBondType())]
                rxn_center[change_idx] = bond_type.join(sorted([bond.GetBeginAtom().GetSymbol(), bond.GetEndAtom().GetSymbol()]))
                change_idx += 1
    return rxn_center, change_bonds_map, change_bonds_idx

check_change_aromatic(rxn)
```

```python
import rdkit 
from rdkit import Chem 

bond_map_type = { '-' : Chem.BondType.SINGLE, '=' : Chem.BondType.DOUBLE, '#' : Chem.BondType.TRIPLE, '@': Chem.BondType.AROMATIC}
bond_elec_cnt = {Chem.BondType.SINGLE: 1, Chem.BondType.DOUBLE: 2, Chem.BondType.TRIPLE: 3, Chem.BondType.AROMATIC: 1.5}
charge_map_inv = {'_': -1, '0': 0, '+': 1, '+2': 2}

def atom_in_ring(mol, atom, ring_size = 6):
    sssr = Chem.GetSymmSSSR(mol)
    atom_idx = atom.GetIdx()
    for ring in sssr:
        if len(ring) == ring_size and atom_idx in ring:
            return True
    return False

def three_degree_N(mol, edit_idx):
    N_list = []
    if type(edit_idx) == type(1):
        check_idx = [edit_idx]
    else:
        check_idx = edit_idx
    for atom in mol.GetAtoms():
        idx = atom.GetIdx()
        if atom.GetIdx() in check_idx:
            if atom.GetSymbol() == 'N':
                used_electron = len(atom.GetNeighbors()) + atom.GetNumExplicitHs() - atom.GetFormalCharge()
                if  atom.GetIsAromatic() and used_electron > 2:
                    N_list.append(atom.GetAtomMapNum())
    return N_list

def correct_electron(mol, atom, force_H = None, N_list = []):
    if force_H is not None:
        atom.SetNoImplicit(True)
        atom.SetNumExplicitHs(force_H)
    atom.SetNoImplicit(True)
    electron = 0
    atom_max_electron = {'C': 4, 'Si': 4, 'Sn': 4, 'N': 3, 'P': 3, 'B': 3, 'O': 2, 'Zn': 0, 'S': 2, 'F': 1, 'Cl': 1, 'Br': 1, 'I': 1, 'Cu': 1, 'Mg': 0, }
    for bond in atom.GetBonds():
        if atom.GetAtomMapNum() in N_list and bond.GetBondType() == Chem.BondType.AROMATIC:
            electron += 1
        else:
            electron += bond_elec_cnt[bond.GetBondType()]
    left_electron = atom_max_electron[atom.GetSymbol()] - electron + atom.GetFormalCharge()
    left_electron = int(left_electron)
    if left_electron >= 0:
        atom.SetNumExplicitHs(left_electron)
    else:
        atom.SetNumExplicitHs(0)
    return 

def dearomatic(mol):
    for bond in mol.GetBonds():
        if not bond.IsInRing() and bond.GetIsAromatic():
            bond.SetBondType(Chem.BondType.SINGLE)
    for atom in mol.GetAtoms():
        if not atom.IsInRing() and atom.GetIsAromatic():
            atom.SetIsAromatic(False)
            origianl_Hs = atom.GetNumExplicitHs()
            correct_electron(mol, atom, origianl_Hs)
    return mol

def add_atom_num(mol):
    for atom in mol.GetAtoms():
        atom.SetAtomMapNum(atom.GetIdx())
    return mol
    
def add_charges(m):
    m.UpdatePropertyCache(strict=False)
    ps = Chem.DetectChemistryProblems(m)
    if not ps:
#         Chem.SanitizeMol(m)
        return m
    for p in ps:
        if p.GetType()=='AtomValenceException':
            at = m.GetAtomWithIdx(p.GetAtomIdx())
            if at.GetAtomicNum()==7 and at.GetFormalCharge()==0 and at.GetExplicitValence()==4:
                at.SetFormalCharge(1)
            elif at.GetAtomicNum()==7 and  at.GetExplicitValence()==4:
                at.SetFormalCharge(1)
            if at.GetAtomicNum()==8 and at.GetFormalCharge()==0 and at.GetExplicitValence()==1:
                at.SetFormalCharge(-1)
#     Chem.SanitizeMol(m)
    return m

def clean_atommap(mol, ignore_atom_num = True, atom_num = False):
    for atom in mol.GetAtoms():
        if ignore_atom_num == False:
            atom.SetAtomMapNum(0)
        elif atom.GetAtomMapNum() != 8 and atom.GetAtomMapNum() != 9:
            atom.SetAtomMapNum(0)
    if atom_num:
        mol = add_atom_num(mol)
#     mol = dearomatic(mol)
    mol = add_charges(mol)
    mol = Chem.MolFromSmiles(Chem.MolToSmiles(mol))
#     return Chem.MolToSmiles(mol, isomericSmiles= False)
    return Chem.MolToSmiles(mol)

def self_connect_LG_bond(product, react_idx, template):
    product_mol = Chem.MolFromSmiles(product)
    last_atom = template[-2]
    for atom in product_mol.GetAtoms():
        if atom.GetIdx() in react_idx and atom.GetSymbol() != last_atom:
            atom.SetAtomMapNum(1)
        elif atom.GetIdx() in react_idx and atom.GetSymbol() == last_atom:
            atom.SetAtomMapNum(2)
        else:
            atom.SetAtomMapNum(0)
            
    change_type = bond_map_type[template[6]]
    LG = template.split('-')[-2]
    LG_mol = Chem.MolFromSmiles(LG)
    
    for atom in LG_mol.GetAtoms():
        if atom.GetAtomMapNum() != 0 and atom.GetSymbol() == last_atom:
            atom.SetAtomMapNum(1)
        elif atom.GetAtomMapNum() != 0 and atom.GetSymbol() != last_atom:
            atom.SetAtomMapNum(2)
            
    combined_mol = Chem.CombineMols(product_mol, LG_mol)
    CO_connect = []
    OC_connect = []
    for atom in combined_mol.GetAtoms():
        if atom.GetAtomMapNum() == 1: 
            CO_connect.append(atom.GetIdx())
        elif atom.GetAtomMapNum() == 2: 
            OC_connect.append(atom.GetIdx())
    emol = Chem.EditableMol(combined_mol)
    # Change bond
    emol.RemoveBond(react_idx[0], react_idx[1]) 
    emol.AddBond(react_idx[0], react_idx[1], change_type)
    # Add Bond
    emol.AddBond(CO_connect[0], CO_connect[1], Chem.BondType.SINGLE) # cut bond
    emol.AddBond(OC_connect[0], OC_connect[1], Chem.BondType.SINGLE) # cut bond
    reactant_mol = emol.GetMol()
    
    for atom_idx in CO_connect + OC_connect:
        atom = reactant_mol.GetAtomWithIdx(atom_idx)
        correct_electron(reactant_mol, atom)
    return clean_atommap(reactant_mol, True)

def self_connect_LG_atom(product, react_idx, template):
    product_mol = Chem.MolFromSmiles(product)
    last_atom = template[-2]
    for atom in product_mol.GetAtoms():
        if atom.GetIdx() == react_idx and atom.GetSymbol() == last_atom:
            atom.SetAtomMapNum(1)
        else:
            atom.SetAtomMapNum(0)       
    LG = template.split('-')[-2]
    LG_mol = Chem.MolFromSmiles(LG)
    for atom in LG_mol.GetAtoms():
        if atom.GetAtomMapNum() != 0 and atom.GetSymbol() == 'C':
            atom.SetAtomMapNum(1)
            
    combined_mol = Chem.CombineMols(product_mol, LG_mol)
    NC_connect = []
    for atom in combined_mol.GetAtoms():
        if atom.GetAtomMapNum() == 1: # CO connect
            NC_connect.append(atom.GetIdx())
    emol = Chem.EditableMol(combined_mol)
    # Add Bond
    emol.AddBond(NC_connect[0], NC_connect[1], Chem.BondType.SINGLE)
    emol.AddBond(NC_connect[0], NC_connect[2], Chem.BondType.SINGLE)
    reactant_mol = emol.GetMol()
    
    for atom_idx in NC_connect:
        atom = reactant_mol.GetAtomWithIdx(atom_idx)
        correct_electron(reactant_mol, atom)
    return clean_atommap(reactant_mol, True)

def search_change_bond(mol, atom, connect_atom, bond_type):
    symbol = connect_atom.GetSymbol()
    charge = connect_atom.GetFormalCharge()
    for n_atom in atom.GetNeighbors():
        bond = mol.GetBondBetweenAtoms(atom.GetIdx(), n_atom.GetIdx())
        if n_atom.GetSymbol() == symbol and bond.GetBondType() != bond_type:
            if n_atom.GetFormalCharge() != charge:
                n_atom.SetFormalCharge(charge)
            return bond
        
def connect_synthon_LG(edit_list, ignore_atom_num = True, one_mol = False):
    if len(edit_list) == 4:
        LG_mol = edit_list[0]
        connect_bond = edit_list[1]
        combined_mol = edit_list[-1]
        if connect_bond == 'NOCHANGE':
            return clean_atommap(combined_mol, ignore_atom_num)
        connect_idx = []
        if edit_list[2] == True:
            combined_mol = Chem.CombineMols(combined_mol, LG_mol)
        for atom in combined_mol.GetAtoms():
            
            if atom.GetAtomMapNum() != 0 and atom.GetAtomMapNum() != 11 and atom.GetAtomMapNum() != 12:
                connect_idx.append(atom.GetIdx())
                
        emol = Chem.EditableMol(combined_mol) 
        if edit_list[2] == True:
            emol.AddBond(connect_idx[0], connect_idx[1], connect_bond) # cut bond
        reactant_mol = emol.GetMol()
        if edit_list[2] == False:
            connect_atom = LG_mol.GetAtomWithIdx(0)
            bond = search_change_bond(reactant_mol, reactant_mol.GetAtomWithIdx(connect_idx[0]), connect_atom, connect_bond)
            bond.SetBondType(connect_bond)
        for atom_idx in connect_idx:
            atom = reactant_mol.GetAtomWithIdx(atom_idx)
            correct_electron(reactant_mol, atom)
            
        return clean_atommap(reactant_mol, ignore_atom_num)
    
    elif len(edit_list) == 7:
        LG_mol1 = edit_list[0]
        connect_bond1 = edit_list[1]
        LG_mol2 = edit_list[3]
        connect_bond2 = edit_list[4]
        combined_mol = edit_list[-1]  #synthon
        connect_idx = []
        change_bond_idx = {}
        synthon_connect_site = 0
            
        for atom in combined_mol.GetAtoms():
            if atom.GetAtomMapNum() != 0:
                synthon_connect_site += 1
                
        if connect_bond1 != 'NOCHANGE' and edit_list[2] == True:
            combined_mol = Chem.CombineMols(combined_mol, LG_mol1)
        if connect_bond2 != 'NOCHANGE' and edit_list[5] == True:
            combined_mol = Chem.CombineMols(combined_mol, LG_mol2)
        
        not_connect_num = [0, 11, 12]
        change_bond_num = []
#         if one_mol:
        if edit_list[2] == False:
            change_bond_num.append(1)
            if one_mol:
                not_connect_num.append(1)
        elif edit_list[5] == False:
            change_bond_num.append(2)
            if one_mol:
                not_connect_num.append(2)
                
        for atom in combined_mol.GetAtoms():
            if atom.GetAtomMapNum() not in not_connect_num:
                connect_idx.append(atom.GetIdx())
            if atom.GetAtomMapNum() in change_bond_num:
                change_bond_idx[atom.GetAtomMapNum()] = atom.GetIdx()
                
#         print (len(connect_idx))
        emol = Chem.EditableMol(combined_mol)
        if connect_bond1 != 'NOCHANGE':
            if edit_list[2] == True:
                if len(connect_idx) == 4: # two atom connect two LG
                    emol.AddBond(connect_idx[0], connect_idx[2], connect_bond1) # add bond
                else:
                    emol.AddBond(connect_idx[0], connect_idx[1], connect_bond1) # add bond
        if connect_bond2 != 'NOCHANGE':
            if edit_list[5] == True:
                if len(connect_idx) >= 3:
                    if one_mol: # two atom connect two LG
                        emol.AddBond(connect_idx[1], connect_idx[3], connect_bond2) # add bond
                    else: # one atom connect two LG
                        emol.AddBond(connect_idx[0], connect_idx[2], connect_bond2) # add bond
                else:
                    emol.AddBond(connect_idx[0], connect_idx[1], connect_bond2) # add bond
        reactant_mol = emol.GetMol()
#         print (Chem.MolToSmiles(reactant_mol))  
        if edit_list[2] == False:
            connect_atom = LG_mol1.GetAtomWithIdx(0)
            if one_mol:
                bond = search_change_bond(reactant_mol, reactant_mol.GetAtomWithIdx(change_bond_idx[1]), connect_atom, connect_bond1)
            else:
                change_idx = list(change_bond_idx.keys())[0]
                bond = search_change_bond(reactant_mol, reactant_mol.GetAtomWithIdx(change_idx), connect_atom, connect_bond1)
            bond.SetBondType(connect_bond1)
            
        if edit_list[5] == False:
            connect_atom = LG_mol2.GetAtomWithIdx(0)
            if one_mol:
                bond = search_change_bond(reactant_mol, reactant_mol.GetAtomWithIdx(change_bond_idx[2]), connect_atom, connect_bond2)
            else:
                change_idx = list(change_bond_idx.keys())[0]
                bond = search_change_bond(reactant_mol, reactant_mol.GetAtomWithIdx(change_idx), connect_atom, connect_bond1)
            bond.SetBondType(connect_bond2)
        
        for atom_idx in connect_idx:
            atom = reactant_mol.GetAtomWithIdx(atom_idx)
            correct_electron(reactant_mol, atom)
            
        for atom_num in change_bond_idx.keys():
            atom = reactant_mol.GetAtomWithIdx(change_bond_idx[atom_num])
            correct_electron(reactant_mol, atom)
            
        return clean_atommap(reactant_mol, ignore_atom_num)
    else:
        LG_mol1 = edit_list[0]
        connect_bond1 = edit_list[1]
        LG_mol2 = edit_list[3]
        connect_bond2 = edit_list[4]
        LG_mol3 = edit_list[6]
        connect_bond3 = edit_list[7]
        combined_mol = edit_list[-1] 
        connect_idx = []
        if connect_bond1 != 'NOCHANGE':
            if edit_list[2] == True:
                combined_mol = Chem.CombineMols(combined_mol, LG_mol1)
        if connect_bond2 != 'NOCHANGE':
            if edit_list[5] == True:
                combined_mol = Chem.CombineMols(combined_mol, LG_mol2)
        if connect_bond3 != 'NOCHANGE':
            if edit_list[8] == True:
                combined_mol = Chem.CombineMols(combined_mol, LG_mol3)
                
        for atom in combined_mol.GetAtoms():
            if atom.GetAtomMapNum() != 0 and atom.GetAtomMapNum() != 11 and atom.GetAtomMapNum() != 12:
                connect_idx.append(atom.GetIdx())
                
        emol = Chem.EditableMol(combined_mol)
        if connect_bond1 != 'NOCHANGE':
            if edit_list[2] == True:
                emol.AddBond(connect_idx[0], connect_idx[1], connect_bond1) # add bond    
        if connect_bond2 != 'NOCHANGE':
            if edit_list[5] == True:
                if len(connect_idx) >= 3:
                    emol.AddBond(connect_idx[0], connect_idx[2], connect_bond2) # add bond
                else:
                    emol.AddBond(connect_idx[0], connect_idx[1], connect_bond2) # add bond
                    
        if connect_bond3 != 'NOCHANGE':
            if edit_list[8] == True:
                if len(connect_idx) >= 4:
                    emol.AddBond(connect_idx[0], connect_idx[3], connect_bond3) # add bond
                elif len(connect_idx) >= 3:
                    emol.AddBond(connect_idx[0], connect_idx[2], connect_bond3) # add bond
                else:
                    emol.AddBond(connect_idx[0], connect_idx[1], connect_bond3) # add bond
                    
        reactant_mol = emol.GetMol()
        
        if edit_list[2] == False:
            connect_atom = LG_mol1.GetAtomWithIdx(0)
            bond = search_change_bond(reactant_mol, reactant_mol.GetAtomWithIdx(connect_idx[0]), connect_atom, connect_bond1)
            bond.SetBondType(connect_bond1)
        if edit_list[5] == False:
            connect_atom = LG_mol2.GetAtomWithIdx(0)
            bond = search_change_bond(reactant_mol, reactant_mol.GetAtomWithIdx(connect_idx[0]), connect_atom, connect_bond2)
            bond.SetBondType(connect_bond2)
        if edit_list[8] == False:
            connect_atom = LG_mol3.GetAtomWithIdx(0)
            bond = search_change_bond(reactant_mol, reactant_mol.GetAtomWithIdx(connect_idx[0]), connect_atom, connect_bond3)
            bond.SetBondType(connect_bond3)
        for atom_idx in connect_idx:
            atom = reactant_mol.GetAtomWithIdx(atom_idx)
            correct_electron(reactant_mol, atom)
            
        return clean_atommap(reactant_mol, ignore_atom_num)

def parse_LG(LG, add_LG = True):
    if len(LG) >=3:
        if ''.join(LG[-3:]) == '-N+' or ''.join(LG[-3:]) == '-O+':
            LG = ''.join(LG[:-1])
            lg_bond_symbol = [bond for bond in LG if bond in bond_map_type.keys()][0]
            lg_bond = bond_map_type[lg_bond_symbol]
            lg_smiles = LG.split(lg_bond_symbol)[1]
            lg_mol = Chem.MolFromSmiles(lg_smiles)
            return lg_mol, lg_bond, add_LG
    if LG[-1] == '+': # need to add levaing group
        LG = LG = ''.join(LG[:-1])
        lg_bond_symbol = [bond for bond in LG if bond in bond_map_type.keys()][0]
        lg_bond = bond_map_type[lg_bond_symbol]
        lg_smiles = lg_bond_symbol.join(LG.split(lg_bond_symbol)[1:])
        lg_mol = Chem.MolFromSmiles(lg_smiles)
        return lg_mol, lg_bond, add_LG
    else: # no need to add anything
        lg_bond_symbol = [bond for bond in LG if bond in bond_map_type.keys()]
        if len(lg_bond_symbol) == 0: # nothing change
            lg_bond = 'NOCHANGE'
            lg_smiles = LG
            lg_mol = Chem.MolFromSmiles(lg_smiles)
            return lg_mol, lg_bond, add_LG
        else: # change bond type
            lg_bond = bond_map_type[lg_bond_symbol[0]]
            lg_smiles = ''.join(LG.split(lg_bond_symbol[0])[1:])
            if '[' not in lg_smiles:
                lg_mol = Chem.MolFromSmiles(lg_smiles)
            else:
                charge = charge_map_inv[''.join(lg_smiles[2:-1])]
                lg_smiles = [a for a in lg_smiles if a.isalpha()][0]
                lg_mol = Chem.MolFromSmiles(lg_smiles)
                lg_mol.GetAtoms()[0].SetFormalCharge(charge)
#             for atom in lg_mol.GetAtoms():
#                 atom.SetAtomMapNum(1)
            return lg_mol, lg_bond, False

def remove_NOCHANGE(mol, LG_mols): # if the synthon atom has no change, assign atom map num 11 and 12
    remove_num = []
    for i, lg_mol in enumerate(LG_mols):
        if lg_mol[1] == 'NOCHANGE':
            remove_num.append(i+1)
    for atom in mol.GetAtoms():
        if atom.GetAtomMapNum() in remove_num:
            if atom.GetAtomMapNum() == 1 or atom.GetAtomMapNum() == 8:
                atom.SetAtomMapNum(11)
            else:
                atom.SetAtomMapNum(12)
    return mol

def change_bond_to_LG(product, react_idx, template):
    LG_smiles = template.split('-')[-1]
    LG_mol = Chem.MolFromSmiles(LG_smiles)
    product_mol = Chem.MolFromSmiles(product)
    for atom in product_mol.GetAtoms():
        if atom.GetIdx() in react_idx and atom.GetSymbol() == 'C': # only leave C
            atom.SetAtomMapNum(1)
        elif atom.GetIdx() in react_idx and atom.GetSymbol() == 'N':
            atom.SetAtomMapNum(2)
        else:
            atom.SetAtomMapNum(0)
    
    connect_idx = []
    combined_mol = Chem.CombineMols(product_mol, LG_mol)
    for atom in combined_mol.GetAtoms():
        if atom.GetAtomMapNum() == 1:
            connect_idx.append(atom.GetIdx())
            
    emol = Chem.EditableMol(combined_mol)
    emol.RemoveBond(react_idx[0], react_idx[1])
    emol.AddBond(connect_idx[0], connect_idx[1], Chem.BondType.SINGLE)
    reactant_mol = emol.GetMol()

    for atom_idx in list(react_idx) + connect_idx: # bond template
        atom = reactant_mol.GetAtomWithIdx(atom_idx)
        correct_electron(reactant_mol, atom)
        
    for atom in reactant_mol.GetAtoms():
        if atom.GetAtomMapNum() == 2:
            remove_idx = atom.GetIdx()  
            
    emol = Chem.EditableMol(reactant_mol)
    emol.RemoveAtom(remove_idx)
    reactant_mol = emol.GetMol()
    return clean_atommap(reactant_mol)

def change_P_to_LG(product, react_idx, template):
    LG_smiles = template.split('+')[0].split('-')[-1] # LG connect on Carbon
    LG_mol = Chem.MolFromSmiles(LG_smiles)
    product_mol = Chem.MolFromSmiles(product)
    # mark edit atom
    for atom in product_mol.GetAtoms():
        if atom.GetIdx() == react_idx[0]: # Carbon (C)
            atom.SetAtomMapNum(1)
        elif atom.GetIdx() == react_idx[1]: # Phosphorus (P)
            atom.SetAtomMapNum(2)
        else:
            atom.SetAtomMapNum(0)
    # cut bond and add halide on carbon
    connect_idx = []
    combined_mol = Chem.CombineMols(product_mol, LG_mol)
    for atom in combined_mol.GetAtoms():
        if atom.GetAtomMapNum() == 1:
            connect_idx.append(atom.GetIdx())  
    emol = Chem.EditableMol(combined_mol)  
    emol.RemoveBond(react_idx[0], react_idx[1]) # cut bond 
    emol.AddBond(connect_idx[0], connect_idx[1], Chem.BondType.SINGLE) # add bond
    reactant_mol = emol.GetMol()
    for atom_idx in list(react_idx) + connect_idx: 
        atom = reactant_mol.GetAtomWithIdx(atom_idx)
        correct_electron(reactant_mol, atom)
    C_smiles = [PG for PG in Chem.MolToSmiles(reactant_mol).split('.') if 'P' not in PG][0]
    
    # fix P-O group
    # mark idx
    P_smiles = [PG for PG in Chem.MolToSmiles(reactant_mol).split('.') if 'P' in PG][0]
    P_mol = Chem.MolFromSmiles(P_smiles)
    marked_O = False
    for atom in P_mol.GetAtoms():
        if atom.GetAtomMapNum() == 2:
            P_idx = atom.GetIdx()
            for n_atom in atom.GetNeighbors():
                bond = P_mol.GetBondBetweenAtoms(atom.GetIdx(), n_atom.GetIdx())
                if bond.GetBondType() == Chem.BondType.SINGLE and marked_O == False:
                    O_add_idx = n_atom.GetIdx()
                    
                    marked_O = True
                elif bond.GetBondType() == Chem.BondType.DOUBLE:
                    n_atom.SetAtomMapNum(3)
    # derive O-R group (ORG)
    emol = Chem.EditableMol(P_mol)
    emol.RemoveBond(P_idx, O_add_idx)
    ORG = emol.GetMol()
    for atom_idx in [P_idx, O_add_idx]:   
        atom = ORG.GetAtomWithIdx(atom_idx)
        if atom.GetSymbol() == 'O':
            atom.SetAtomMapNum(2)
        correct_electron(ORG, atom)
    ORG_smiles = [org for org in Chem.MolToSmiles(ORG).split('.') if 'P' not in org][0]
    ORG_mol = Chem.MolFromSmiles(ORG_smiles)
    # remove P=O and attach ORG
    connect_idx = []
    combined_mol = Chem.CombineMols(P_mol, ORG_mol)
    for atom in combined_mol.GetAtoms():
        if atom.GetAtomMapNum() == 2:
            connect_idx.append(atom.GetIdx())
        elif atom.GetAtomMapNum() == 3:
            O_remove_idx = atom.GetIdx()
    emol = Chem.EditableMol(combined_mol)
    emol.RemoveBond(connect_idx[0], O_remove_idx)
    emol.AddBond(connect_idx[0], connect_idx[1], Chem.BondType.SINGLE) # add bond
    emol.RemoveAtom(O_remove_idx)
    P_mol = emol.GetMol()
    clean_idx = []
    for atom in P_mol.GetAtoms():
        if atom.GetAtomMapNum() == 2:
            clean_idx.append(atom.GetIdx())
    for atom_idx in clean_idx: 
        atom = P_mol.GetAtomWithIdx(atom_idx)
        correct_electron(P_mol, atom, force_H = 0)
    
    reactant_list = sorted([clean_atommap(P_mol), clean_atommap(Chem.MolFromSmiles(C_smiles))])
    return '.'.join(reactant_list)

def apply_tempalte(product, react_idx, template):
    if ''.join(template[-3:]) == '-N+' and type(react_idx) == type(1):
        return self_connect_LG_atom(product, react_idx, template)
    elif ''.join(template[-3:]) == '-O+':
        return self_connect_LG_bond(product, react_idx, template)
    elif template == 'C-N->C-N[CH:1]=O':
        return change_bond_to_LG(product, react_idx, template)
    elif 'P-OC' in template:
        return change_P_to_LG(product, react_idx, template)

    mol = Chem.MolFromSmiles(product)
      
    if type(react_idx) == type(1): # atom template
        edit_type = 'atom'
        for atom in mol.GetAtoms():
            if atom.GetIdx() == react_idx:
                atom.SetAtomMapNum(1)
            else:
                atom.SetAtomMapNum(0)
    else:  # bond template
        edit_type = 'bond'
        for atom in mol.GetAtoms():
            if atom.GetIdx() == react_idx[0]:
                atom.SetAtomMapNum(1)
            elif atom.GetIdx() == react_idx[1]:
                atom.SetAtomMapNum(2)
            else:
                atom.SetAtomMapNum(0)
    
    LGs = template.split('->')[1].split('$')
    if len(LGs) == 2:
        LGs[1].replace(':1', ':2')
    
    change_type = None
    if len(LGs) == 3: # have bond change and att LG
        for atom in mol.GetAtoms():
            if atom.GetIdx() == react_idx[0]:
                atom.SetAtomMapNum(8)
            elif atom.GetIdx() == react_idx[1]:
                atom.SetAtomMapNum(9)
            else:
                atom.SetAtomMapNum(0)
        change_type = bond_map_type[LGs[0][1]]
        LGs = LGs[1:]
    
    LG_mols = []
    for i, LG in enumerate(LGs):
        edit_info = []
        if LG[0] == '[': # need to change formal charge
            charge_word = ''.join(str(LG.split(']')[0])[2:])
            change_charge = charge_map_inv[charge_word]
            if edit_type == 'atom':
                mol.GetAtomWithIdx(react_idx).SetFormalCharge(change_charge)
            else:
                mol.GetAtomWithIdx(react_idx[i]).SetFormalCharge(change_charge)
            
        if '&' in LG:
            lgs = LG.split('&')
            for lg in lgs:
                lg_mol, lg_bond, LG_bool = parse_LG(lg)
                edit_info.append(lg_mol)
                edit_info.append(lg_bond)
                edit_info.append(LG_bool)
        else:        
            lg_mol, lg_bond, LG_bool = parse_LG(LG)
            edit_info.append(lg_mol)
            edit_info.append(lg_bond)
            edit_info.append(LG_bool)
        LG_mols.append(edit_info)
    
    mol = remove_NOCHANGE(mol, LG_mols) # mark NOCHANGE atom
    N_list = three_degree_N(mol, react_idx)
    
    if '$' in template:
        cut_bond_type = mol.GetBondBetweenAtoms(react_idx[0], react_idx[1]).GetBondType()
    else:
        cut_bond_type = 'None'
        
    if list(template).count('$') == 2 and len(template) == 12: # pure bond change (bond template)
        change_bond_type = bond_map_type[template[6]]
        mol.GetBondBetweenAtoms(react_idx[0], react_idx[1]).SetBondType(change_bond_type)
        for atom_idx in react_idx:
            correct_electron(mol, mol.GetAtomWithIdx(atom_idx))
        reactant = clean_atommap(mol, False)
        if template == 'C-N->C=N$C$N':
            print ('Change stereo')
            reactant = add_stereo(reactant, react_idx)
        return reactant
    
    elif '$' not in template and '+' not in template: # pure bond change (atom template)
        change_bond_type = bond_map_type[template[6]]
        mol.GetBondBetweenAtoms(react_idx[0], react_idx[1]).SetBondType(change_bond_type)
        for atom_idx in react_idx:
            correct_electron(mol, mol.GetAtomWithIdx(atom_idx))
        return clean_atommap(mol, False)

    emol = Chem.EditableMol(mol)  
    if '$' in template: # need to cut a bond
        emol.RemoveBond(react_idx[0], react_idx[1]) # cut bond 
    cut_mol = emol.GetMol()
    
    if str(cut_bond_type) == 'AROMATIC': # cut aromatic bond
        cut_mol = dearomatic(cut_mol)
    if type(react_idx) == type(1): # atom template
        atom = cut_mol.GetAtomWithIdx(react_idx)
        correct_electron(cut_mol, atom, N_list = N_list)
    else:
        for atom_idx in react_idx: # bond template
            atom = cut_mol.GetAtomWithIdx(atom_idx)
            if atom.GetFormalCharge() != 0: # reset formal charge
                atom.SetFormalCharge(0)
            if atom.GetSymbol() == 'P':
                atom.SetFormalCharge(1)
            correct_electron(cut_mol, atom, N_list = N_list)
            
    if template == 'C-C->C$C-CS(=O)(=O)[OH:1]+':
        atom = cut_mol.GetAtomWithIdx(react_idx[0])
        atom.SetFormalCharge(-1)
        correct_electron(cut_mol, atom)
        
                
    for synthon in Chem.MolToSmiles(cut_mol).split('.'):
        if len(LG_mols) == 2 and len(Chem.MolToSmiles(cut_mol).split('.')) == 1: # cut bond but not split
            LG_mols = LG_mols[0] + LG_mols[1]
            LG_mols.append(Chem.MolFromSmiles(synthon))
            LG_mols = [LG_mols]
        elif ':1]' in synthon or ':8]' in synthon or ':11]' in synthon: # cut bond and split # 1
            LG_mols[0].append(Chem.MolFromSmiles(synthon))
        elif len(LG_mols) == 2: # cut bond and split # 2
            LG_mols[1].append(Chem.MolFromSmiles(synthon))
    
#     LG_mols = [lg_mol for lg_mol in LG_mols if len(lg_mol) in [4, 7, 10]]
#     print (LG_mols)
#     print (Chem.MolToSmiles(cut_mol))
    if len(LG_mols) == 1 and len(LG_mols[0]) == 7 and '&' not in template: # One molecule editing
        reactant_list = sorted([connect_synthon_LG(edit_list, one_mol = True) for edit_list in LG_mols])
    else:
        reactant_list = sorted([connect_synthon_LG(edit_list) for edit_list in LG_mols])
    if change_type == None:
        reactants = final_clean('.'.join(reactant_list))
        return reactants
    else:
#         print ('SSSSSSSSSSSSSSSEEEEEEEEEEECCCCCCCCCCCCOOOOOOOOOONNNNNNNNDDDDDDD', '.'.join(reactant_list))
        second_step = [Chem.MolFromSmiles(reactant_list[0]), change_type, True, Chem.MolFromSmiles(reactant_list[1])]
        return connect_synthon_LG(second_step, False)                
                    
def final_clean(smiles):
    if 'C#N' in smiles.split('.'):
        print (smiles.split('.')[0] == 'C#N')
        if smiles.split('.')[0] == 'C#N':
            smiles = '.'.join(sorted([smiles.split('.')[1], '[C-]#N']))
        else:
            smiles = '.'.join(sorted([smiles.split('.')[0], '[C-]#N']))
    
    mol = Chem.MolFromSmiles(smiles)
    for atom in mol.GetAtoms():
        atom.SetAtomMapNum(0)
    return Chem.MolToSmiles(mol)
 
def add_stereo(smi, react_idx):
    m = Chem.MolFromSmiles(smi)
    bid = m.GetBondBetweenAtoms(react_idx[0], react_idx[1]).GetIdx()
    m.GetBondWithIdx(bid).SetStereoAtoms(react_idx[0], react_idx[1])
    m.GetBondWithIdx(bid).SetStereo(Chem.BondStereo.STEREOTRANS)
    return Chem.MolToSmiles(m)

reaction_center = (0, 1)
rxn = 'C1CO[CH:2]([c:3]2[cH:4][cH:5][c:6]([C:7](=[O:8])[c:9]3[cH:10][n:11]([CH2:12][c:13]4[cH:14][cH:15][cH:16][c:17]([Br:18])[n:19]4)[c:20]4[cH:21][cH:22][cH:23][cH:24][c:25]4[c:26]3=[O:27])[cH:28][cH:29]2)[O:1]C1>>[O:1]=[CH:2][c:3]1[cH:4][cH:5][c:6]([C:7](=[O:8])[c:9]2[cH:10][n:11]([CH2:12][c:13]3[cH:14][cH:15][cH:16][c:17]([Br:18])[n:19]3)[c:20]3[cH:21][cH:22][cH:23][cH:24][c:25]3[c:26]2=[O:27])[cH:28][cH:29]1'
product_smiles = rxn.split('>')[-1]
gt = rxn.split('>')[0]
template = 'C=O->C-O$C$C-C(C[OH:1])[CH3:1]-O+'
print (clean_atommap(Chem.MolFromSmiles(gt), False))
pred = apply_tempalte(product_smiles, reaction_center, template)
print (pred)
```
